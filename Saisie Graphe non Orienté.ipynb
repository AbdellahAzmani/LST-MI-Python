```python
# libraries
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt


class Graph(object):

    def __init__(self, graph_dict=None):
        if graph_dict == None:
            graph_dict = {}
        self.__graph_dict = graph_dict

    def vertices(self):
        return list(self.__graph_dict.keys())

    def edges(self):
        return self.__generate_edges()

    def add_vertex(self, vertex):
        if vertex not in self.__graph_dict:
            self.__graph_dict[vertex] = []

    def add_edge(self, edge):
        edge = set(edge)
        (vertex1, vertex2) = tuple(edge)
        if vertex1 in self.__graph_dict:
            self.__graph_dict[vertex1].append(vertex2)
        else:
            self.__graph_dict[vertex1] = [vertex2]
    
    def __generate_edges(self):
        edges = []
        for vertex in self.__graph_dict:
            for neighbour in self.__graph_dict[vertex]:
                if {neighbour, vertex} not in edges:
                    edges.append({vertex, neighbour})
        return edges

    def __str__(self):
        res = "vertices: "
        for k in self.__graph_dict:
            res += str(k) + " "
        res += "\nedges: "
        for edge in self.__generate_edges():
            res += str(edge) + " "
        return res

    def vertex_degree(self, vertex):
            adj_vertices =  self.__graph_dict[vertex]
            degree = len(adj_vertices) + adj_vertices.count(vertex)
            
    def toMatrice(self):
        for i in self.__graph_dict:
            print()
            for j in self.__graph_dict:
                if i in self.__graph_dict[j]:
                    print ('1', end=" ")
                else:
                    print ('0', end=" ")
    
if __name__ == "__main__":

    g = {}
    
    graph = Graph(g)  
    
    while True:
        try:
            graph.add_vertex(input("saisir sommet : ")) 
            i = int(input("------Tapez votre choix------(0 pour exit)\n"))
            if i ==0:
                break   
        except Exception as e:
            print("e")
    
    print('--------------------------------')        
    
    while True:
        try:
            graph.add_edge([input("saisir sommet 1 : "),input("saisir sommet 2 : ")])
            k = int(input("------Tapez votre choix------(0 pour exit)\n"))
            if k ==0:
                break   
        except Exception as e:
            print("e")     

    print('--------------------------------')
    print("Les sommets du graphe : ")
    print(graph.vertices())
    print('--------------------------------')
    print("Les arêtes du graphe : ")
    print(graph.edges())
    print('--------------------------------')
    print( 'La matrice d\'adjacence de G est :' )
    A = graph.toMatrice()
    print('--------------------------------')
    print( 'Degré des sommets :' )
    for node in graph.vertices():
        print(graph.vertex_degree(node))
    print('--------------------------------')
    print("Dessin du graphe non orinté ")
    h = nx.DiGraph(g)
    nx.draw(h, with_labels=True)
```

```python
# Dessin du graphe non orinté
h = nx.DiGraph(g)
nx.draw(h, with_labels=True)
```

```python
# Parcours en profondeur
print("Parcours en Profondeur")
print("Liste des chemins suivis")
print("------------------------")
print(list(nx.dfs_edges(g, source="D")))

print("\nReprésentation sous forme d'arbre")
print("---------------------------------")
tree = nx.dfs_tree(g, source="D")
nx.draw(tree, with_labels=True, pos=nx.spring_layout(g))
```

```python
# Parcours en largeur

print("Parcours en Largeur")
print("Liste des chemins suivis")
print("------------------------")
print(list(nx.bfs_edges(h, source="D")))

print("\nReprésentation sous forme d'arbre")
print("---------------------------------")
tree = nx.bfs_tree(h, source="D")
nx.draw(tree, with_labels=True, pos=nx.spring_layout(h))
```

```python
print("Algorithme de Dijkstra")
print("------------------------")
A = nx.dijkstra_path(h, "D", "B")
print(A)
```

```python
print("Algorithme de Bellman-Ford")
print("------------------------")
print(nx.bellman_ford_path(h, "A", "B"))
```
