### Nommage

Le nom d’une variable ne peut être composé que de lettres, majuscules ou minuscules, de chiffres et du symbole souligné « _ » (celui de la touche « 8 »).


<center style="color:red">**ChiffreAffaires, Chiffre_Affaire, TVA, TauxValeurAjoutee, Taux_Valeur_Ajoutee**</center>


Un nom de variable ne doit pas commencer par un chiffre.
<center style="color:red">**« 7jours » ou « 7_jours »**</center>
ne sont pas considérées comme variables.

 Le langage Python est sensible à la casse (différence entre Majuscule et Minuscule) : 
<center style="color:red">**«note», «Note» et  «NOTE» sont trois variables différentes **</center>
 Un nom de variable ne peut comprendre les caractères suivants : « espace », « apostrophe », « guillement », ni caractères spéciaux …
<center style="color:red">**‘ | " |  ~ | # | (| ) |  [ | ] | & | { } |  + |  - |  * |  /  **</center>

| Exemples nommage correct | Exemples nommage incorrects |
| :--------- | :----------------------- |
| Premier_age | 1_age |
| Nombre_Enfants | Nombre_d’enfants |
| Formule1 | Formule 1 |
| Moyenne_Generale | Moyenne_Générale |

### Typage des variables

Contrairement au langage classique C, Pascal, C++, Java, …, il n’est pas nécessaire de déclarer vos variables car Python comprend automatiquement de quel type est une variable dès son affectation.

Comme il est pratique de pouvoir savoir de quel type est une variable, Python propose la fonction prédéfinie « type ». 

Il suffit de l’appeler par : type(nom_variable)

| Entier | Réel | Chaîne de Caractères |
| :-- | :-- | :-- |
| >>> a = 3 | >>> b = 0.25  | >>> d = "Informatique"|
| >>> type(a) | >>> type(b) | >>> type(d) |
| <class ‘int'> | <class 'float'> | <class ‘str'> |

## Expression arithmétique

### Opérateurs et expressions

**Addition, soustraction, multiplication, division**

Pour effectuer ces opérations, on utilise respectivement les symboles +, -, *, / et **.

```python
a = 3
a = a + 3
b = a - 2
c = a * b
d = c / 10
e = (a-b)**(b-1)
print("a = ", a, "  b = ", b, "  c = ", c, "  d = ", d, " e = ", e )
```

**Division entière et modulo**

Il existe deux autres opérateurs qui permettent de connaître le résultat d'une division entière et le reste de cette division.

Le premier opérateur utilise le symbole « // ». Il permet d'obtenir la partie entière d'une division.

L'opérateur « % », que l'on appelle le « modulo », permet de connaître le reste de la division.

```python
a = 15
b = a / 10
c = a // 10
d = a % 10
print("a = ", a, "  b = ", b, "  c = ", c, "  d = ", d )
```

**Incrémentation**

L'opérateur+=revient à ajouter à la variable la valeur qui suit l'opérateur. 

Les opérateurs -=, *= et /= existent également, bien qu'ils soient moins utilisés.

```python
a = 15
a += 7
print("1. a = ", a )
```

```python
a -= 2
print(" 2. a = ", a )
```

```python
a /= 4
print(" 3. a = ", a )
```

```python
a *= 3
print(" 4. a = ", a )
```

## Expression Arithmétique

### Priorités des opérateurs :

Priorités des opérateurs :
Pour respecter la priorité des opérateurs on utilise  un mot  mnémotechnique qui correspond à l’acronyme PEMDAS :

P pour parenthèses. Ce sont elles qui ont la plus haute priorité. Elles vous permettent donc de « forcer » l’évaluation d’une expression dans l’ordre que vous voulez :  
<center style="color:red">2*(3-1) = 4.</center> 

E pour exposants. Les exposants sont évalués ensuite, avant les autres opérations.
<center style="color:red">2**1+1 = 3 (et non 4).</center>

M et D pour multiplication et division, qui ont la même priorité. Elles sont évaluées avant l’addition A et la soustraction S, lesquelles sont donc effectuées en dernier lieu. 
<center style="color:red">	2*3-1 = 5   (plutôt que 4), </center>
<center style="color:red">	et 2/3-1 = -0.3333...  (Plutôt que 1.0).</center>

Si deux opérateurs ont la même priorité, l’évaluation est effectuée de gauche à droite.
<center style="color:red"> **59*100//60**</center>

* la multiplication est effectuée en premier => 5900
* puis la division 5900//60
* ce qui donne 98. 

<center style="color:red"> **Si la division était effectuée en premier, le résultat serait 59 **</center>

**Exemple 1 :**

« 5 + 30 * 20 » est équivalent à « 5 + (30 * 20) » soit 605

« 5 + 30 * 20 / 10 » équivaut à « 5 + ((30 * 20) / 10) » soit 65

5 + 600 / 10<br>
5 + 60<br>
65
 
**Exemple 2 :**

7 * 2 + (7 + 3 * (5-2)) / 4 * 2<br>
7 * 2 + 16 / 4 * 2     #  G -> D<br>
14 + 4 * 2<br>
14 + 8<br>
22
