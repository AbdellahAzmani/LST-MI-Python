## Graphes non orientés
### Définition :

Un graphe G est un couple (V,E) où 

V = {v1, v2, ..., vn} est un ensemble fini dont les éléments sont appelés sommets. 
E = {e1, e2, ..., em} est un ensemble fini dont les éléments sont appelés arêtes. 

Une arête e de l’ensemble E est définie par une paire non-ordonnée (x,y) de sommets, appelés les extrémités de e

Si l'arête e relie les sommets a et b, on dira que ces sommets sont adjacents , ou encore que l'arête e est incidente avec les sommets a et b.

On appelle ordre d’un graphe le nombre de sommets (n) de ce graphe 

### Exemple :

Ensemble des sommets: 
V = {1, 2, 3, 4, 5}     
                                                  
Ensemble des arêtes: 
E = {(1, 3), (1, 4), (1, 5), (2, 3), (3, 4), (3, 5), (4, 5)}

### Graphe partiel:

Soit G = (V, E) un graphe. Le graphe G' = (V, E') est un graphe partiel de G, si E' est inclus dans E
 
### Sous graphe :

Le graphe G’=(A, E(A)) est un sous graphe de G, si A est un sous ensemble de V et E(A) est l'ensemble des arêtes formé de toutes les arêtes de G ayant leurs deux extrémités dans A 

* Un graphe partiel d'un sous-graphe est un sous-graphe partiel de G
* On appelle clique un sous-graphe complet de G 
* On appelle stable un sous-graphe de G sans arêtes

### Degré d'un sommet :

Pour un graphe ou un multigraphe, on appelle degré du sommet v, et on note d(v),

## Graphes orientés
### Définitions :

Un digraphe fini G = (V, E) est défini par l'ensemble fini V = {v1, v2, ..., vn} dont les éléments sont appelés sommets, et par l'ensemble fini E = {e1, e2, ..., em}  dont les éléments sont appelés arcs. 

Un arc e de l'ensemble E est défini par une paire ordonnée de sommets 
e = (u, v), on dira que l'arc e va de u à v. 

* On note d+(v) le degré extérieur du sommet v
* On note d-(v) le degré intérieur du sommet v 

Le degré d'un sommet orienté est la somme de degré  extérieur et de degré intérieur : d(v) = d+(v) + d-(v )

## Chemin et cycle 

* Un **chemin** est une liste p=(x1,x2,…., xk) de sommets telle qu'il existe dans le graphe une arête entre chaque paire de sommets successifs. 
* La **longueur** du chemin correspond au nombre d'arêtes parcourues : k-1 
* Un chemin p est **simple** si chaque arête du chemin est visité une seule fois. 
* Un **cycle**  est un chemin simple finissant à son point de départ : x1=xk+1
* Un chemin p=(x1,x2,...,xk) est **élémentaire** si chacun des sommets du parcours est visité une seule fois 
* Un chemin **élémentaire** est donc un chemin simple et sans cycle
* Un graphe est **connexe** ssi il existe un chemin entre chaque paire de sommets.

## Graphes eulériens et hamiltoniens 

* Un graphe est **eulérien** s'il est possible de trouver un cycle passant une et une seule fois par toutes les arêtes. 
* Un graphe est **semi-eulérien** s'il est possible de trouver un chemin passant une et une seule fois par toutes les arêtes
* Un graphe est **hamiltonien** s'il est possible de trouver un cycle passant une et une seule fois par tous les sommets
* Un graphe est **semi-hamiltonien** s'il est possible de trouver un chemin passant une et une seule fois par tous les sommets.

### Théorème 
Soit G = (V, E) un graphe simple d'ordre n >= 3. Si pour toute paire {x,y} de sommets non adjacents, on a d(x) + d(y) >= n, alors G est hamiltonien.

## Arbres
### Définitions :
* On appelle arbre tout graphe connexe, sans cycles. 
* Un graphe sans cycles et non connexe est appelé une forêt.

### Théorème 
Les affirmations suivantes sont équivalentes pour tout  graphe G = (V, E) ayant n sommets. 
1. G est un arbe, 
2. G est sans cycles et connexe, 
3. G est sans cycles et comporte n-1 arrêtes, 
4. G est connexe  et comporte n-1 arrêtes, 
5. Chaque paire {u, v} de  sommets distincts est reliée par un seul chemin simple (et le graphe est sans boucles).

##  Réseaux de Transport et flots 
### Réseau de transport :
Un réseau de transport est un graphe orienté, connexe, sans boucle et valué, où chaque arc est associé à un nombre c(u) , appelé "capacité" de l'arc u. 
* Il existe un seul noeud s arbitraire appelé l'entrée du réseau, ou la source. 
* Il existe également un seul noeud t arbitraire appelé la sortie du réseau. 
* Entre les noeuds s et t, il peut exister des noeuds intermédiaires. 

<img src="graphes/20.png"/>

### Définition :
Un flot f dans un réseau de transport est une fonction qui associe à chaque arc u une quantité f(u) qui représente la quantité de flot qui passe par cet arc, en provenance de la source et en destination du puits.

### Loi de Kirchhoff : 
La somme des intensités qui arrivent au noeud est égale à la somme des intensités qui en repartent.
