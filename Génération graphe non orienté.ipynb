## Génération d'un graphe non Orienté

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

G = nx.random_geometric_graph(8, 0.6)
A=nx.MultiGraph(G)

cities = {0:"A",1:"B",2:"C",3:"D",4:"E",5:"F",6:"G",7:"H"}

H=nx.relabel_nodes(G,cities)

nx.draw(H, with_labels=True)
plt.show()
```

```python
print("Nodes of graph: ")
print(H.nodes())
print("Edges of graph: ")
print(H.edges())
```

```python
nx.to_numpy_matrix(H)
```

```python
print("Liste des chemins suivis")
print("------------------------")
print(list(nx.dfs_edges(H, source="A")))

print("\nReprésentation sous forme d'arbre")
print("---------------------------------")
tree = nx.dfs_tree(H, source="A")
nx.draw(tree, with_labels=True, pos=nx.spring_layout(H))
```

```python
print("Liste des chemins suivis")
print("------------------------")
print(list(nx.bfs_edges(H, source="A")))

print("\nReprésentation sous forme d'arbre")
print("---------------------------------")
tree = nx.bfs_tree(H, source="A")
nx.draw(tree, with_labels=True, pos=nx.spring_layout(H))
```

```python
print("Algorithme de Dijkstra")
print("------------------------")
A = nx.dijkstra_path(H, "A", "E")
print(A)
```

```python
print("Algorithme de Bellman-Ford")
print("------------------------")
print(nx.bellman_ford_path(H, "A", "B"))
```

```python
print("Vérifier si le graphe est orienté")
print("------------------------")
nx.is_directed(H)
```
