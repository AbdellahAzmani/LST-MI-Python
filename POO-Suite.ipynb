### Exercice 1 :
Ecrire les fonctions qui permettent de calculer l'expression ensembliste suivante par application de la loi de Morgan :

(A \/ B)  /\ (A \/ C) \/ (A \/ B)  /\ (A \/ C)  


#### Pour rappel, les Règles de distributivité de Morgan sont : 
     A /\ (B \/ C) = (A /\ B) \/ (A /\ C)     
     A \/ (B /\ C) = (A \/ B) /\ (A \/ C)

     (A \/ B)  /\ (A \/ C) \/ (A \/ B)  /\ (A \/ C) = A \/ (B /\ C)

```python
def intersection(A,B):
    C=[]
    for i in range(len(A)):
        if (A[i] in B):
            C=C+[A[i]]
    return C
        
def union(A,B):
    C=B
    for i in range(len(A)):
        if (A[i] not in B):
            C=C+[A[i]]
    return C

def morgan(A,B,C):
    return(union(A,intersection(B,C)))  
```

### Exercice 2 :
Ecrire la fonction Python qui correspond à l'algorithme de la suite Fibonacci.

#### Pour rappel, la suite de Fibonacci est : 
    F(0)=F(1)=1 et F(n)=F(n-1)+F(n-2) pour n>1

```python
def Fibo1(N): #Version récursive
    if (N==0)or(N==1):
        return(1)
    else:
        return(Fibo1(N-1)+Fibo1(N-2))

def Fibo2(N): #Version itérative
    if (N==0)or(N==1):
        return(1)
    else:
        x=1
        y=1
        i=2
        while i<=N:
            z=x+y
            x=y
            y=z
            i+=1
        return(z)
```
