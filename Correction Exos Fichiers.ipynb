```python
# Exo 9
#9. Recréer le fichier « paysMonde » à partir des 3 fichiers précédents (attention à la redondance) en indiquant  dans chaque ligne le « Continent : pays »
# 1. On va utiliser le principe de l'exo 3. Créer la Liste de caractères retournée par la lecture de ce fichier
# On crée 3 listes et on fait leur union pour avoir une nouvelle liste sans doublons
# 2. On va utiliser le principe de l'exercice
# on crée un nouveau fichier à partir de la nouvelle liste
#***************************************************************************************************************************
def Fichier_Vers_Liste(Fich):
    # ouvrir le fichier en mode lecture 
    monFichier=open(Fich, 'r') 
    # copier le contenu du fichier dans une liste de caractères
    Liste=monFichier.readlines()
    monFichier.close()
    return (Liste)

#***************************************************************************************************************************
def union(L1,L2):
    L3=L2
    for i in range(len(L1)):
        if (not(L1[i]in L3)):
            L3=L3+[L1[i]]
    return(L3)

#***************************************************************************************************************************
def Liste_Vers_Fichier(List, Fic):
    # Ouverture du fichier en mode écriture « w »
    monFichier=open(Fic, 'w') 
    # Ecriture de la liste en utilisant la fonction « writelines() »
    monFichier.writelines(List)  
    # Fermeture du fichier avec la méthode close().
    monFichier.close()  
#***************************************************************************************************************************
def Afficher_Fichier(File):
    # Affichage du fichier avec l'instruction "with"
    with open(File, 'r') as Fichier:
        for ligne in Fichier:
            print (ligne)      
#***************************************************************************************************************************
def Exo_9():
    L1=Fichier_Vers_Liste('/content/drive/My Drive/PaysAfrique.txt')
    L2=Fichier_Vers_Liste('/content/drive/My Drive/PaysArabes.txt')
    L3=Fichier_Vers_Liste('/content/drive/My Drive/PaysEuropeens.txt')
    Liste_Vers_Fichier(union(L1,union(L2,L3)),"/content/drive/My Drive/PaysMonde.txt")
    Afficher_Fichier("/content/drive/My Drive/PaysMonde.txt")
```
