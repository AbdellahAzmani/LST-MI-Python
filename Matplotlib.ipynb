## Matplotlib
### Qu'est-ce que Matplotlib?
* Matplotlib est une bibliothèque de tracé de graphiques de bas niveau en python qui sert d'utilitaire de visualisation.
* Matplotlib a été créé par John D. Hunter.
* Matplotlib est open source et nous pouvons l'utiliser librement.
* Matplotlib est principalement écrit en python, quelques segments sont écrits en C, Objective-C et Javascript pour la compatibilité avec la plate-forme.

### Où est la base de code Matplotlib?
Le code source de Matplotlib se trouve dans ce référentiel github https://github.com/matplotlib/matplotlib

### Import Matplotlib

```python
import matplotlib
```

## Pyplot
La plupart des utilitaires Matplotlib se trouvent sous le sous-module **pyplot**, et sont généralement importés sous l'alias **plt**:

```python
import matplotlib.pyplot as plt
```

#### Tracer une ligne dans un diagramme de la position (0,0) à la position (6,250):

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

### Tracer les points x et y
* La fonction **plot()** est utilisée pour dessiner des points (marqueurs) dans un diagramme.
* Par défaut, la fonction **plot()** trace une ligne d'un point à l'autre.
* La fonction prend des paramètres pour spécifier des points dans le diagramme.
* Le paramètre 1 est un tableau contenant les points sur l'axe des x.
* Le paramètre 2 est un tableau contenant les points sur l'axe y.

Si nous devons tracer une ligne de (1, 3) à (8, 10), nous devons passer deux tableaux [1, 8] et [3, 10] à la fonction plot.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### Tracer sans ligne
Pour tracer uniquement les marqueurs, vous pouvez utiliser le paramètre de notation de chaîne de raccourci «o», qui signifie «anneaux».

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

### Points multiples
Vous pouvez tracer autant de points que vous le souhaitez, assurez-vous simplement d'avoir le même nombre de points sur les deux axes.

#### Exemple
Tracez une ligne dans un diagramme de la position (1, 3) à (2, 8) puis à (6, 1) et enfin à la position (8, 10):

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### Points X par défaut
Si nous ne spécifions pas les points sur l'axe des x, ils obtiendront les valeurs par défaut 0, 1, 2, 3, etc. en fonction de la longueur des points y.

Donc, si nous prenons le même exemple que ci-dessus et omettons les points x, le diagramme ressemblera à ceci:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

## Marqueurs
Vous pouvez utiliser le marqueur d'argument de mot-clé pour souligner chaque point avec un marqueur spécifié:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

### Formater les lignes fmt
Vous pouvez également utiliser le paramètre de notation de chaîne de raccourci pour spécifier le marqueur.

Ce paramètre est également appelé **fmt** et est écrit avec cette syntaxe:
* marqueur
* ligne
* couleur

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

#### Référence de ligne
* '-'	Solid line
* ':'	Dotted line
* '--'	Dashed line
* '-.'	Dashed/dotted line

#### Référence de couleur
* 'r'	Red	
* 'g'	Green
* 'b'	Blue
* 'c'	Cyan
* 'm'	Magenta
* 'y'	Yellow
* 'k'	Black
* 'w'	White

### Taille du marqueur 
Vous pouvez utiliser le mot-clé **markersize** ou la version plus courte **ms** pour définir la taille des marqueurs:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

### Couleur du marqueur
Vous pouvez utiliser l'argument mot-clé **markeredgecolor** ou le **mec** plus court pour définir la couleur du bord des marqueurs:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

### Ligne
Vous pouvez utiliser le mot-clé **linestyle**, ou **ls** plus court, pour changer le style de la ligne tracée:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

#### Syntaxe plus courte
Le style de ligne peut être écrit dans une syntaxe plus courte:

* linestyle : ls.
* dotted : .
* dashed : --


#### Couleur de la ligne
Vous pouvez utiliser le mot-clé **color** ou le plus court **c** pour définir la couleur de la ligne:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r')
plt.show()
```

### Titre
Vous pouvez ajouter un titre à chaque tracé avec la fonction **title()**:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Vente")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Achat")

plt.show()
```

## Création de nuages de points
Avec Pyplot, vous pouvez utiliser la fonction **scatter()** pour dessiner un nuage de points.

La fonction **scatter()** trace un point pour chaque observation. Il a besoin de deux tableaux de même longueur, un pour les valeurs de l'axe des x et un pour les valeurs de l'axe des y:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

## Création des barres

## Création des histogrammes
Un histogramme est un graphique montrant les distributions de fréquences.

Il s'agit d'un graphique montrant le nombre d'observations dans chaque intervalle donné.

Dans Matplotlib, nous utilisons la fonction **hist()** pour créer des histogrammes.

La fonction **hist()** utilisera un tableau de nombres pour créer un histogramme, le tableau est envoyé dans la fonction en tant qu'argument.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

## Création de graphiques à secteurs
Avec Pyplot, vous pouvez utiliser la fonction **pie()** pour dessiner des graphiques à secteurs:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

### Étiquettes
Ajoutez des étiquettes au graphique à secteurs avec le paramètre **label**.

Le paramètre **label** doit être un tableau avec une étiquette pour chaque wedge:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

### Angle de départ
Comme mentionné, l'angle de départ par défaut est sur l'axe des x, mais vous pouvez modifier l'angle de départ en spécifiant un paramètre d'angle de départ.

Le paramètre startangle est défini avec un angle en degrés, l'angle par défaut est 0:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

### Exploser
Peut-être voulez-vous que l'un des coins se démarque? Le paramètre **explode** vous permet de faire cela.

Le paramètre **explode**, s'il est spécifié, et **None**, doit être un tableau avec une valeur pour chaque coin.

Chaque valeur représente à quelle distance du centre chaque coin est affiché:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

### Légende
Pour ajouter une liste d'explications pour chaque coin, utilisez la fonction **legend()**:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

## Visualisation des  Fonctions trigonométriques
En combinant Numpy et Matplotlib, nous avons la possibilité de visualiser les différentes fonctions.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4,4,100)
y = np.cos(x)+3*np.sin(2*x)
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def circle(coords, r):
    X, Y = [], []
    for t in range(101):
        X.append(coords[0]+r*np.cos(t*np.pi/50))
        Y.append(coords[1]+r*np.sin(t*np.pi/50))
        plt.plot(X, Y, 'b')
```
